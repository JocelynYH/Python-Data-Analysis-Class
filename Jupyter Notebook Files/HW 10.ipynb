```python
# Background: The folder Stocks.zip contains nearly 300 files, each including daily data for
# a specific stock, with stock ticker symbol given in the file name. Each observation includes 
# the following:
#      Date = date information recorded
#      Open = opening stock price
#      High = high stock price
#      Low = low stock price
#      Close = closing stock price
#      Volume = number of shares traded
#      Adj Close = closing price adjusted for stock splits (ignored for this assignment)
# The time interval covered varies from stock to stock. There are some missing records, so the
# data is incomplete. Note that some dates are not present because the exchange is closed on 
# weekends and holidays. Those are not missing records. Answer the questions below based on 
# the data available in the files.

# 1. Find the mean for the Open, High, Low, and Close entries for all of the records taken as 
# a single group.
```

```python
import pandas as pd
import numpy as np
import glob



filelist = glob.glob('*.csv') 

df = pd.DataFrame()

tickers = []
for i in range(len(filelist)):
    tickers.append(filelist[i].split('.csv')[0])

tickers

x=0
for f in filelist:
    newdf = pd.read_csv(f)
    newdf['ticker'] = tickers[x]
    x += 1
    df = pd.concat([df,newdf])

df

nums = df.loc[:,'Open':'Close']
nums.mean(axis=0)
'''
#1
Open     50.863852
High     51.459412
Low      50.253368
Close    50.876482
dtype: float64
'''
```

```python
# 2. Find the top-5 and bottom-5 stocks in terms of their average Close price. Give tables 
# showing the stock ticker symbol and the average Close price. (Print out of a Series or 
# DataFrame is fine.)
closebyticker = df['Close'].groupby(df['ticker'])  # group by the ticker id & grab stock's close prices
sortedcbt = closebyticker.mean().sort_values()
print("The cheapest 5 stocks in terms of average Close price are", sortedcbt.head(5))
print("The top 5 stocks with the highest Close price are", sortedcbt.tail(5))
'''
#2
The cheapest 5 stocks in terms of average Close price are ticker
FTR      8.969515
F       11.174158
XRX     11.291864
ETFC    12.808103
HBAN    13.697483
Name: Close, dtype: float64

The top 5 stocks with the highest Close price are ticker
GS      139.146781
BLK     164.069088
AMZN    185.140534
AZO     235.951950
CME     253.956017
Name: Close, dtype: float64'''
```

```python
# 3. Find the top-5 and bottom-5 stocks in terms of the day-to-day volatility of the price. 
# This is the standard deviation of the differences High - Low daily for each stock. (Use the 
# sample standard deviation.) Give tables for each, as in the previous problem.

df['diff'] = df['High'] - df['Low'] # compute difference between high and low prices and store in column
diffbyticker = df['diff'].groupby(df['ticker']) # grab differences in prices by ticker name
volatility = diffbyticker.std().sort_values()  # compute standard deviations by ticker and sort
print(volatility.head(5))  # subset and print lowest 5
print(volatility.tail(5))  # subset and print highest 5

'''
#3
ticker
FTR     0.141359
NWSA    0.174096
XRX     0.176320
F       0.194724
NI      0.195203
Name: diff, dtype: float64
ticker
BLK     3.716916
NFLX    4.283600
AZO     4.288179
AMZN    4.410448
CME     8.040071
Name: diff, dtype: float64
'''
```

```python
# 4. Repeat the previous problem, this time using the relative volatility, which is given by
df['openyclose'] = df['Open'] + df['Close']
df['calculation'] = ((df['diff'])/(0.5*df['openyclose']))
calculationbyticker = df['calculation'].groupby(df['ticker']) # grab differences in prices by ticker name
sortedbyrv = calculationbyticker.mean().sort_values()
print(sortedbyrv.head(5))
print(sortedbyrv.tail(5))
'''
#4
ticker
GIS    0.007912
WEC    0.009308
K      0.009378
CL     0.009520
ZTS    0.009861
Name: calculation, dtype: float64
ticker
AAL     0.043955
LVLT    0.045748
EQIX    0.053003
AIG     0.053350
GGP     0.065728
Name: calculation, dtype: float64
'''
```

```python
# 5. For each day the market was open in February 2010, find the average daily price for all 
# stocks for each of Open, High, Low, Close, and Volume.

df['date'] = pd.to_datetime(df['Date']) 
feb10 = df[(df['date'].dt.month == 2) & (df['date'].dt.year == 2010)]
feb10.loc[:,'Open':'Volume'].mean(axis=0)
'''
#5
Open      4.269530e+01
High      4.320367e+01
Low       4.220083e+01
Close     4.280406e+01
Volume    7.416517e+06
dtype: float64
'''
```

```python
# 6. For 2012, find the date with the maximum average relative
# volatility for all stocks and the date with the minimum average 
# relative volatility for all stocks. (Consider only days when the 
# market is open.)

year2012 = df[(df['date'].dt.year == 2012)] # subset by year
year2012
```

```python
sortRV = year2012.groupby(year2012['date']).mean() # subset the 2012 entries by date and calculate their means
RVbyDate = sortRV.sort_values(by='calculation')  # find the means of their relative volatilities and sort
print(RVbyDate.head(1))  # the lowest avg relative volatility
print(RVbyDate.tail(1))  # the highest avg relative volatility

'''
#6
The date with the maximum average relative volatility is 2012-12-24.  The date with the minimum average relative volatility is 2012-06-21. 
'''
```
